```python
# Ejercicio 3: Red neuronal en el dataset Iris
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Cargar dataset Iris
iris = load_iris()
X = iris.data
y = iris.target

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Crear modelo
mlp = MLPClassifier(hidden_layer_sizes=(10, 15, 20, 30), activation="relu", max_iter=2000)

#Entrenar
mlp.fit(X_train, y_train)

#Predecir
y_pred = mlp.predict(X_test)
print("Precision en test: ", accuracy_score(y_test, y_pred))

#TAREA: CARGAR DATASET ALEATORIO, Y PROBAR Y VER CON CUANTAS NEURONAS FUNCIONA Y ES ÓPTIMO.
```
